2) Jelaskan apa yang dimaksud dengan unsupervised learning
    Jawab :
    Unsupervised learning adalah salah satu paradigma dalam machine learning di mana model diberikan data tanpa label atau informasi target yang jelas. Dalam unsupervised learning, tujuan utamanya adalah menemukan pola atau struktur dalam data tanpa memiliki petunjuk eksplisit tentang output yang diharapkan. Model unsupervised learning diberi tugas untuk menjelajahi data dan menemukan hubungan atau pola yang tersembunyi tanpa panduan yang spesifik.
    
    Terdapat dua jenis tugas utama dalam unsupervised learning:
    a. **Clustering (Pengelompokan):**
   - Tujuan dari tugas clustering adalah mengelompokkan data ke dalam kelompok-kelompok yang serupa berdasarkan kesamaan fitur atau karakteristik tertentu.
   - Algoritma seperti K-Means, Hierarchical Clustering, dan DBSCAN adalah contoh algoritma clustering yang umum digunakan.
   
   b. **Dimensionality Reduction (Reduksi Dimensi):**
   - Tujuan dari tugas reduksi dimensi adalah mengurangi jumlah fitur atau dimensi dalam data tanpa kehilangan informasi penting. Ini membantu mengatasi masalah "kutub tinggi" (curse of dimensionality) dan memfasilitasi pemahaman yang lebih baik terhadap struktur data.
   - Algoritma seperti Principal Component Analysis (PCA), t-SNE, dan Autoencoders adalah contoh algoritma reduksi dimensi.
   
   Contoh aplikasi dari unsupervised learning termasuk:
   - **Segmentasi Pelanggan:** Mengelompokkan pelanggan berdasarkan pola pembelian atau perilaku yang serupa.
   - **Deteksi Anomali:** Mengidentifikasi data yang berbeda atau anomali dalam suatu dataset.
   - **Ekstraksi Fitur:** Menemukan fitur-fitur penting dalam data yang dapat digunakan untuk analisis lebih lanjut.
   - **Kompresi Data:** Mengurangi dimensi data untuk mengurangi kompleksitas dan kebutuhan komputasi.
   
   Unsupervised learning berbeda dengan supervised learning, di mana model dilatih dengan data yang memiliki label yang menunjukkan output yang diinginkan. Dalam unsupervised learning, model harus mengidentifikasi pola atau struktur sendiri tanpa bimbingan dari data label.
